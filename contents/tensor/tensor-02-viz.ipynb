# Visulized Tensor

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import torch
```

## Matplotlib

```python
# Create sample tensors
tensor_1d = torch.randn(100)
tensor_2d = torch.randn(28, 28)  # Like MNIST image
tensor_3d = torch.randn(3, 64, 64)  # RGB image
```

```python
# 1D tensor - line plot
plt.figure(figsize=(10, 3))
plt.plot(tensor_1d.numpy())
plt.title("1D Tensor Visualization")
plt.show()
```

```python
# 2D tensor - heatmap/image
plt.figure(figsize=(6, 6))
plt.imshow(tensor_2d.numpy(), cmap='gray')
plt.colorbar()
plt.title("2D Tensor as Grayscale Image")
plt.show()
```

```python
# 3D tensor - RGB image (channels-first format)
plt.figure(figsize=(6, 6))
# Convert from CHW to HWC format for matplotlib
img = tensor_3d.permute(1, 2, 0)
# Normalize to [0,1] range
img = (img - img.min()) / (img.max() - img.min())
plt.imshow(img.numpy())
plt.title("3D Tensor as RGB Image")
plt.show()
```
